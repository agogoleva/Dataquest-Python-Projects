<a id = 2> Clean the data: </a> - [Top](#0)

We analyse the two columns containing answers on the following questions:

- Have you seen any of the 6 films in the Star Wars franchise?
- Do you consider yourself to be a fan of the Star Wars film franchise?

There are 3 possible values: Yes / No / NaN.

To work with this data, it is useful to convert it to Boolean values True / False / NaN, respectively.

```python
#use a dictionary
yes_no = {"Yes": True,
         "No": False}

#call the map() function to perform the mapping
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)

star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)

star_wars.head()
```

```python
star_wars.columns[3:9]
import re
#Extract movie name containing in columns 3 to 8 to a list
movies = [str(star_wars.iloc[:,i].dropna().unique()).replace("[", "").replace("]", "").replace("'", "") for i in range(3, 9)]

#Create a disctionary
movies_mapping = {}
for movie in movies:
    movies_mapping[movie] = True
movies_mapping[np.nan] = False
print(movies_mapping)

#Reaplce manes of the movies with boolean expressions in columns 3 to 8
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movies_mapping)    
```

```python
#Rename columns 
new_names1 = ["seen_1","seen_2","seen_3","seen_4","seen_5","seen_6"]
column_indices1 = [3,4,5,6,7,8]
old_names1 = star_wars.columns[column_indices1]
star_wars.rename(columns=dict(zip(old_names1, new_names1)), inplace=True)
star_wars.head()
```

```python
#Convert column 9 up to but not including column 15 to the float data type
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

#Rename columns 
new_names2 = ["ranking_1","ranking_2","ranking_3","ranking_4","ranking_5","ranking_6"]
column_indices2 = [9,10,11,12,13,14]
old_names2 = star_wars.columns[column_indices2]
star_wars.rename(columns=dict(zip(old_names2, new_names2)), inplace=True)
star_wars.head()
```

# Find the highest-ranked movie 

```python
#Mean of each of the ranking columns
#A mean value per movie
print(star_wars[star_wars.columns[9:15]].mean())
#Bar chart
plt.bar(range(1,7),star_wars[star_wars.columns[9:15]].mean())
```

The Star Wars funs rated the 6 episodes from 1 to 6, where 1 is the highest rank. The first 3 episodes have decreasing ranking (from 3.7 to 4.3), giving the lowest ranking to the 3rd episode 'Star Wars: Episode III  Revenge of the Sith'. The following movies have higher ranking, where the 5th episode 'Star Wars: Episode V The Empire Strikes Back' ranked being the top (2.5).

# Finding the most seen movie

```python
#how many people have seen each movie
print(star_wars[star_wars.columns[3:9]].sum())

#Bar chart
plt.bar(range(1,7),star_wars[star_wars.columns[3:9]].sum())
```

According to how many people have seen a certain episode, we can conlude that the original movies were seen by more respondents than the newer movies, where 'Star Wars: Episode V The Empire Strikes Back' is the most viewed episode.

# The most viewed movie and the highest-ranked movie accoding to a fun's gender

```python
#split the dataframe into 2 according to the gender
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
#Mean of each of the ranking columns
#A mean value per movie
plt.bar(range(1,7),females[females .columns[9:15]].mean(),ls='dashed', alpha = 0.1, lw=3, color= 'r')
plt.bar(range(1,7),males[males .columns[9:15]].mean(),ls='dashed', alpha = 0.1, lw=3, color= 'b')
plt.legend(["Female", "Male"])
plt.show()
#how many people have seen each movie
plt.bar(range(1,7),females[females .columns[3:9]].sum(),ls='dashed', alpha = 0.1, lw=3, color= 'r')
plt.bar(range(1,7),males[males .columns[3:9]].sum(),ls='dashed', alpha = 0.1, lw=3, color= 'b')
plt.legend(["Female", "Male"])
plt.show()
```

Male fans seems to appreciate more the original movies, while female fans favourite more the new episodes. Overaall, more male funs have seen each episode compare to the female funs.
